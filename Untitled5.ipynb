```r
# Load necessary libraries
library(tidyverse)  # For data manipulation and visualization
library(caret)      # For model training and evaluation
library(ggplot2)    # For visualization
library(randomForest) # For random forest model
library(dplyr)      # For data manipulation
```

```r
# Step 1: Load the data
titanic_data <- read.csv('titanic.csv')  # Assuming the Titanic dataset is in the working directory
```

```r
#Read the data set
head(titanic_data)
```

```r
# Check for missing values in all columns of the dataset
missing_values <- colSums(is.na(titanic_data))

# Print the number of missing values for each column
print(missing_values)
```

```r
# Step 2: Data Cleaning
summary(titanic_data)

# Handle missing values for Age (replace with median)
titanic_data$Age[is.na(titanic_data$Age)] <- median(titanic_data$Age, na.rm = TRUE)
# Remove rows where 'Embarked' is an empty string
# Find the mode (most frequent value) of the 'Embarked' column
mode_embarked <- names(sort(table(titanic_data$Embarked), decreasing = TRUE))[1]

# Replace empty values in 'Embarked' column with the mode
titanic_data$Embarked[titanic_data$Embarked == ""] <- mode_embarked

# Remove the 'Cabin' column
titanic_data$Cabin <- NULL
```

```r
print(sum(is.na(titanic_data$Age)))
```

```r
# Install and Load the necessary library
install.packages("corrplot")
library(corrplot)

# Select numeric columns for correlation
numeric_cols <- titanic_data[, c("Age", "Fare", "SibSp", "Parch", "Pclass")]

# Calculate correlation matrix
cor_matrix <- cor(numeric_cols, use = "complete.obs")

# Plot the correlation matrix
corrplot(cor_matrix, method = "circle")
```

```r
# Scatterplot matrix (pair plot) to visualize relationships between numerical variables
pairs(titanic_data[, c("Age", "Fare", "SibSp", "Parch", "Pclass")],
      main = "Scatterplot Matrix of Numerical Variables")
```

```r
# Step 3: Data Visualization

# Visualize distribution of Survived vs Pclass
ggplot(titanic_data, aes(x = factor(Pclass), fill = factor(Survived))) +
  geom_bar(position = "fill") +
  labs(title = "Survival rate by Passenger Class", x = "Passenger Class", y = "Survival Proportion") +
  scale_fill_manual(values = c("red", "green"), labels = c("Not Survived", "Survived"))

# Visualize survival rate by Gender
ggplot(titanic_data, aes(x = factor(Sex), fill = factor(Survived))) +
  geom_bar(position = "fill") +
  labs(title = "Survival rate by Gender", x = "Gender", y = "Survival Proportion") +
  scale_fill_manual(values = c("red", "green"), labels = c("Not Survived", "Survived"))
```

```r
# Scatter plot: Age vs Fare with Survival status indicated by color
ggplot(titanic_data, aes(x = Age, y = Fare, color = factor(Survived))) +
  geom_point() +
  labs(title = "Scatter plot of Age vs Fare by Survival Status",
       x = "Age",
       y = "Fare") +
  scale_color_manual(values = c("red", "green"), labels = c("Not Survived", "Survived")) +
  theme_minimal()
```

```r
# Pie chart: Proportion of Survival
survival_count <- table(titanic_data$Survived)
survival_data <- as.data.frame(survival_count)
names(survival_data) <- c("Survived", "Count")

ggplot(survival_data, aes(x = "", y = Count, fill = factor(Survived))) +
  geom_bar(stat = "identity", width = 1) +
  coord_polar(theta = "y") +
  scale_fill_manual(values = c("red", "green"), labels = c("Not Survived", "Survived")) +
  labs(title = "Survival Distribution", fill = "Survival Status") +
  theme_void()
```

```r
# Get unique values for both 'Pclass' and 'Embarked'
unique_values <- list(
  Pclass = unique(titanic_data$Pclass),
  Sex = unique(titanic_data$Sex),
  Embarked = unique(titanic_data$Embarked)
)

# Print the unique values
print(unique_values)
```

```r
# Step 4: Feature Extraction
# Convert categorical variables to factors
titanic_data$Survived <- factor(titanic_data$Survived, levels = c(0, 1))
titanic_data$Pclass <- factor(titanic_data$Pclass)
titanic_data$Sex <- factor(titanic_data$Sex)
titanic_data$Embarked <- factor(titanic_data$Embarked)

# Check the first few rows of the dataset to confirm changes
head(titanic_data)
```

```r
# Get unique values for both 'Pclass' and 'Embarked'
unique_values <- list(
  Pclass = unique(titanic_data$Pclass),
  Sex = unique(titanic_data$Sex),
  Embarked = unique(titanic_data$Embarked)
)

# Print the unique values
print(unique_values)
```

```r
# Step 5: Model Training (Random Forest)

# Split the data into training and testing sets (80/20 split)
set.seed(123)  # Set seed for reproducibility
trainIndex <- createDataPartition(titanic_data$Survived, p = 0.8, list = FALSE)
train_data <- titanic_data[trainIndex, ]
test_data <- titanic_data[-trainIndex, ]

# Train the random forest model
rf_model <- randomForest(Survived ~ Pclass + Sex + Age + SibSp + Parch + Fare + Embarked,
                         data = train_data,
                         ntree = 100)
```

```r
# Step 6: Model Evaluation

# Make predictions on the test set
predictions <- predict(rf_model, newdata = test_data)

# Confusion matrix to evaluate model
conf_matrix <- confusionMatrix(predictions, test_data$Survived)
print(conf_matrix)
```
